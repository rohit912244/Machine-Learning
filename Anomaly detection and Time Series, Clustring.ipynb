Q1. What is Dimensionality Reduction? Why is it important in machine
learning?

Ans: Dimensionality Reduction is the process of reducing the number of input features (dimensions) in a dataset while preserving as much important information as possible.

1. Reduces Overfitting

- Too many features can cause models to memorize noise instead of learning patterns.

- Fewer, more meaningful features help models generalize better.

2. Improves Model Performance

- Fewer dimensions mean:

- Faster training

- Lower memory usage

- Simpler models

- This is especially important for large datasets.

3. Solves the Curse of Dimensionality

- As dimensions increase, data becomes sparse.

- Distance-based algorithms (like KNN, clustering) perform poorly in high dimensions.

- Dimensionality reduction makes these algorithms more effective.

4. Removes Redundant and Irrelevant Features

- Many features may be correlated or unimportant.

- Dimensionality reduction keeps only the most informative ones.

5. Helps with Data Visualization

- Humans can easily visualize data in 2D or 3D.

- Reducing dimensions allows:

- Pattern discovery

- Cluster visualization

- Outlier detection

6. Noise Reduction

- Eliminates noisy features that do not contribute meaningful information.

- Improves signal quality for learning algorithms.

Q2. Name and briefly describe three common dimensionality reduction
techniques.

Ans: Here are three common dimensionality reduction techniques with brief descriptions:

1. Principal Component Analysis (PCA)

- A feature extraction technique.

- Transforms original features into a smaller set of new, uncorrelated variables called principal components.

- Each component captures the maximum possible variance in the data.

- Commonly used for noise reduction and visualization.

2. Linear Discriminant Analysis (LDA)

- A supervised dimensionality reduction technique.

- Maximizes class separability by finding feature combinations that best separate different classes.

- Often used in classification problems.

3. t-Distributed Stochastic Neighbor Embedding (t-SNE)

- A non-linear dimensionality reduction technique.

- Preserves local similarities between data points.

- Mainly used for data visualization in 2D or 3D, especially for high-dimensional data.

Q3. What is clustering in unsupervised learning? Mention three popular
clustering algorithms.

Ans:Clustering is an unsupervised learning technique that groups data points into clusters such that:

- Data points within the same cluster are more similar to each other

- Data points in different clusters are less similar to each other

- No labeled data is required; the algorithm discovers patterns and structure directly from the data.

Popular Clustering Algorithms:
1. K-Means Clustering

- Partitions data into K predefined clusters.

- Assigns each data point to the nearest cluster centroid.

- Efficient and widely used for large datasets.

2. Hierarchical Clustering

- Builds a tree-like structure called a dendrogram.

Can be:

- Agglomerative (bottom-up)

- Divisive (top-down)

- Does not require pre-specifying the number of clusters.

3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

- Groups data based on density.

- Can find clusters of arbitrary shape.

- Identifies noise and outliers effectively.

Q5. List and briefly describe three types of anomaly detection techniques.

Ans: three common types of anomaly detection techniques, with brief descriptions:

1. Statistical-Based Techniques:

These methods assume that normal data follows a known statistical distribution (e.g., Gaussian). Data points that significantly deviate from the expected distribution (such as being beyond a threshold like 3 standard deviations) are considered anomalies.

Example: Z-score, Grubbs’ test.

2. Distance-Based (or Proximity-Based) Techniques:

Anomalies are identified based on their distance from other data points. Points that are far away from most of the data (i.e., have few nearby neighbors) are treated as anomalies.

Example: k-Nearest Neighbors (k-NN) based anomaly detection.

3. Machine Learning–Based Techniques:

These methods learn patterns of normal behavior from data and flag instances that do not conform to the learned model. They can work in supervised, semi-supervised, or unsupervised settings.

Example: Isolation Forest, One-Class SVM, Autoencoders.

Q6. What is time series analysis? Mention two key components of time series
data.

Ans: Time series analysis is the study of data points collected or recorded at regular time intervals to identify patterns, trends, and underlying structures over time, and to make forecasts for future values.

Two key components of time series data are:

1. Trend – The long-term movement or direction in the data (upward, downward, or constant) over time.

2. Seasonality – Regular, repeating patterns or fluctuations that occur at fixed intervals (e.g., daily, monthly, yearly).

Q7. Describe the difference between seasonality and cyclic behavior in time
series.

Ans: Seasonality and cyclic behavior in time series differ mainly in their regularity and predictability:

Seasonality:

Refers to patterns that repeat at fixed and known intervals (such as daily, monthly, quarterly, or yearly). These patterns are usually caused by calendar-related factors like weather, holidays, or business cycles.
Example: Higher retail sales every December.

Cyclic Behavior:

Refers to long-term fluctuations that occur over irregular periods and do not have a fixed or predictable length. Cycles are often influenced by economic or external factors.
Example: Economic expansions and recessions.

Q8.Write Python code to perform K-means clustering on a sample dataset.

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Create a sample dataset
X, y = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Get cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200)
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

Q9. What is inheritance in OOP? Provide a simple example in Python.

Ans: Inheritance in Object-Oriented Programming (OOP) is a mechanism where a child (derived) class acquires the properties and methods of a parent (base) class. This promotes code reusability, supports hierarchical classification, and allows the child class to extend or modify the behavior of the parent class.

```python
# Parent class
class Animal:
    def speak(self):
        print("Animal makes a sound")

# Child class inherits from Animal
class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Create an object of Dog
dog = Dog()

dog.speak()  # Inherited method
dog.bark()   # Child class method
```

Q10. How can time series analysis be used for anomaly detection?

Ans: Time series analysis is used for anomaly detection by learning the normal temporal patterns of data and identifying observations that deviate significantly from those patterns.

Key steps:

Model normal behavior:

Fit a time series model that captures trend, seasonality, and noise (e.g., moving average, ARIMA, exponential smoothing, or ML models).

Forecast expected values:

Use the model to predict expected values (often with confidence intervals) at each time point.

Compare actual vs. expected:

Compute residuals (actual − predicted). Large or unusual residuals indicate potential anomalies.

Flag anomalies:

Points outside thresholds or confidence bounds are labeled as anomalies.

Common techniques:

- Statistical: Z-score, control charts (Shewhart, EWMA)

- Forecasting-based: ARIMA/ETS residual analysis

- Machine learning: Isolation Forest, One-Class SVM, LSTM/Autoencoder models

Use cases: detecting fraud, sensor faults, network intrusions, and unexpected demand spikes.
