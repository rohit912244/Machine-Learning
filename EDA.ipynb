Q4. Create a bar plot to visualize the average selling price for each seller_type and write one observation.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/BikeDetails.csv.csv')

# Check for missing values
print(df.isnull().sum)

# Check column names
print(df.columns)

# Compute average selling price by seller type
avg_price = df.groupby("seller_type")["selling_price"].mean().sort_values()

# Plot bar chart
plt.figure(figsize=(8,5))
avg_price.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Average Selling Price by Seller Type")
plt.xlabel("Seller Type")
plt.ylabel("Average Selling Price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Observation example:
# Individual sellers tend to have lower average selling prices compared to dealers.
```

Q6.Use the IQR method to detect and remove outliers from the km_driven
column. Show before-and-after summary statistics.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/BikeDetails.csv.csv')

# Check for missing values
print(df.isnull().sum)

# Check column names
print(df.columns)

# Display initial summary statistics for 'km_driven'
print("--- Summary Statistics for 'km_driven' BEFORE Outlier Removal ---")
initial_stats = df['km_driven'].describe()
print(initial_stats)

# Calculate Q1, Q3, and IQR
Q1 = df['km_driven'].quantile(0.25) # 13500.0
Q3 = df['km_driven'].quantile(0.75) # 43000.0
IQR = Q3 - Q1 # 43000.0 - 13500.0 = 29500.0

# Define the outlier bounds
# Lower Bound: Q1 - 1.5 * IQR = 13500.0 - 1.5 * 29500.0 = -30750.0
# Upper Bound: Q3 + 1.5 * IQR = 43000.0 + 1.5 * 29500.0 = 87250.0
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers using the IQR method: filter for values between the bounds
df_clean = df[(df['km_driven'] >= lower_bound) & (df['km_driven'] <= upper_bound)]

# Display final summary statistics for 'km_driven'
print("\n--- Summary Statistics for 'km_driven' AFTER Outlier Removal ---")
final_stats = df_clean['km_driven'].describe()
print(final_stats)

# Prepare the combined statistics for display in the response
stats_df = pd.DataFrame({
    'Before Removal': initial_stats,
    'After Removal': final_stats
})

print("\n--- Combined Summary Statistics ---")
print(stats_df)
```

Q9.Generate a heatmap of the correlation matrix for all numeric columns.
What correlations stand out the most?


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('BikeDetails.csv.csv')

# Select the numeric columns for correlation analysis
df_numeric = df[['selling_price', 'year', 'km_driven', 'ex_showroom_price']]

# Calculate the correlation matrix
correlation_matrix = df_numeric.corr()

# Set up the matplotlib figure
plt.figure(figsize=(8, 6))

# Draw the heatmap
sns.heatmap(correlation_matrix,
            annot=True,
            fmt=".2f",
            cmap='coolwarm',
            cbar=True,
            linewidths=.5,
            linecolor='black')

plt.title('Correlation Matrix of Numeric Bike Details Columns', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()

# Save the plot
plt.savefig('correlation_heatmap.png')
```

Q10.Summarize your findings in a brief report:

â— What are the most important factors affecting a bike's selling price?

â— Mention any data cleaning or feature engineering you performed.

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("BikeDetails.csv.csv")
CURRENT_YEAR = 2025

# --- Feature Engineering ---

# 1. Bike Age
df['bike_age'] = CURRENT_YEAR - df['year']

# 2. Bike Brand
df['brand'] = df['name'].apply(lambda x: x.split()[0])

# --- Data Cleaning (for ex_showroom_price correlation) ---

# Create a subset for correlation analysis involving ex_showroom_price
# Drop rows where ex_showroom_price is missing
df_price_clean = df.dropna(subset=['ex_showroom_price']).copy()


# --- Factor Analysis (Correlation) ---

# Correlation on the full dataset for km_driven and bike_age
correlation_full = df[['selling_price', 'km_driven', 'bike_age']].corr().loc['selling_price', ['km_driven', 'bike_age']]

# Correlation including ex_showroom_price (using the clean subset)
correlation_subset = df_price_clean[['selling_price', 'ex_showroom_price']].corr().loc['selling_price', 'ex_showroom_price']

print("--- Correlation of Selling Price with Numerical Features ---")
print(f"km_driven: {correlation_full['km_driven']:.4f}")
print(f"bike_age: {correlation_full['bike_age']:.4f}")
print(f"ex_showroom_price (subset): {correlation_subset:.4f}")

# --- Factor Analysis (Categorical - Mean Selling Price) ---

categorical_factors = ['seller_type', 'owner', 'brand']

print("\n--- Mean Selling Price by Categorical Factor (Top 5 for Brand) ---")
for col in categorical_factors:
    mean_price_by_cat = df.groupby(col)['selling_price'].mean().sort_values(ascending=False)
    if col == 'brand':
        print(f"\nTop 5 {col}s by Mean Selling Price:")
        print(mean_price_by_cat.head(5).to_markdown(numalign="left", stralign="left"))
    else:
        print(f"\n{col} by Mean Selling Price:")
        print(mean_price_by_cat.to_markdown(numalign="left", stralign="left"))
```
