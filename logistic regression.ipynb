Q2. What are the key assumptions of Simple Linear Regression?

Ans:
1. Linearity

- The relationship between the independent variable (X) and dependent variable (Y) should be linear.

That means changes in X lead to proportional changes in Y.

- Check: Scatter plot of X vs. Y should show a roughly straight-line pattern.

2. Independence of Errors

- The residuals (errors) should be independent of each other.

- In time series data, this means that residuals at one time point shouldnâ€™t be correlated with those at another.

- Check: Durbinâ€“Watson test or residual plot.

3. Homoscedasticity (Constant Variance of Errors)

- The variance of the residuals should be constant across all values of X.

- In other words, the spread of residuals should be roughly the same for all predicted values.

- Check: Residuals vs. Fitted Values plot should show a random scatter (no funnel or pattern).

4. Normality of Errors

- The residuals (errors) should be normally distributed.

- This assumption is especially important for hypothesis testing (like t-tests on coefficients).

Check: Histogram or Qâ€“Q plot of residuals.

5. No (or Minimal) Measurement Error in X

- The independent variable (X) should be measured without significant error.

- Errors in X can bias the slope estimation.

 6. No Perfect Multicollinearity (only for multiple regression)

- In Simple Linear Regression, this isnâ€™t applicable because thereâ€™s only one predictor.

- But it becomes important in Multiple Linear Regression (when there are multiple X variables).

Q5. What is the method of least squares in linear regression?

Ans: The Method of Least Squares is a procedure used to find the best-fitting line through a set of data points by minimizing the sum of the squared errors (residuals) between the observed values and the predicted values.

Q7. Name and briefly describe three common evaluation metrics for regression models.

Ans: Three common evaluation metrics for regression models are Mean Absolute Error (MAE), Mean Squared Error (MSE), and Root Mean Squared Error (RMSE). MAE measures the average absolute difference between predicted and actual values, while MSE squares these differences before averaging them. RMSE is the square root of MSE, which brings the metric back to the original units of the target variable.


Q8. What is the purpose of the R-squared metric in regression analysis?

Ans: The purpose of the R-squared metric in regression analysis is to measure how well the independent variables in a model explain the variation in the dependent variable, providing a goodness-of-fit measure for the model. It indicates the percentage of the variance in the dependent variable that is predictable from the independent variables, with a higher value (closer to 100%) showing a better fit.

Â Explains variance: R-squared quantifies the proportion of the variance in the dependent variable that can be accounted for by the independent variables in the model.

Indicates model fit: It serves as a goodness-of-fit statistic, telling you how closely the data points are to the regression line or curve.

Ranges from 0 to 1: The value ranges from \(0\) to \(1\), where \(0\) means the model explains none of the variability, and \(1\) means it explains all of the variability.

Represents a percentage: It can be interpreted as a percentage, so an R-squared of \(0.76\) means that \(76\%\) of the variation in the dependent variable is explained by the model.

Helps compare models: It can be used to compare the explanatory power of different regression models.

Q9. : Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

```python
# Import necessary libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
# X: independent variable (hours studied)
# Y: dependent variable (exam score)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # must be 2D for sklearn
Y = np.array([30, 40, 50, 60, 70])

# Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Print the slope (coefficient) and intercept
print("Slope (Î²1):", model.coef_[0])
print("Intercept (Î²0):", model.intercept_)
```
